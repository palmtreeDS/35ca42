```python
a = sns.load_dataset("iris")
b = sns.FacetGrid(a, col = 'species')
b.map(plt.hist, 'sepal_length')
```

```python
import time
s =time.time()
b = sns.FacetGrid(a, col = 'species')
b.map(plt.hist, 'sepal_width')
e =time.time()
lap = e-s
print(lap)
```

```python
s = time.time()
b = sns.PairGrid(a)
b.map(plt.scatter)
e = time.time()
lap = e-s
print(lap)
```
