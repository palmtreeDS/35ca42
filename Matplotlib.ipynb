```python
style.use('ggplot')
x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

plt.plot(x,y,'g', label = 'line one', linewidth = 5)
plt.plot(x2,y2,'c', label = 'line two', linewidth = 5)

plt.title('Epic Info')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')

plt.legend()
plt.grid(True, color = 'k')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.bar([1,2,5,7,9],[5,2,7,8,2], label = "Example One")
plt.bar([2,4,6,8,10],[8,6,2,5,6],label = "Example Two", color = 'g')

plt.legend()
plt.xlabel('Bar number')
plt.ylabel('Bar Height')
plt.title('Bar Graph')
plt.show()
```

```python
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]
plt.plot([],[],color = 'm',label = 'sleeping', linewidth = 5)
plt.plot([],[],color = 'c',label = 'Eating',linewidth =5)
plt.plot([],[],color = 'r',label = 'working', linewidth = 5)
plt.plot([],[],color = 'k',label = 'playing', linewidth = 5)
plt.stackplot(days, sleeping, eating, working, playing, colors = ['m','c','r','k'])
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Area Plot')
plt.legend()
plt.show()
```

```python
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]
plt.plot([],[],color = 'c',label = 'Eating', linewidth = 5)
plt.plot([],[],color = 'm',label = 'sleeping',linewidth =5)
plt.plot([],[],color = 'r',label = 'working', linewidth = 5)
plt.plot([],[],color = 'k',label = 'playing', linewidth = 5)
plt.stackplot(days, eating, sleeping, working, playing, colors = ['c','m','r','k'])
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Area Plot')
plt.legend()
plt.show()
```

```python
slices = [7,2,2,13]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['c','m','r','b']
plt.pie(slices, labels = activities,
          colors = cols,
        startangle = 90,
        shadow = True,
        explode = (0,0.1,0,0.1),
        autopct = '%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
